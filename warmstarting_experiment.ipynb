# Coil-20 dataset

## Coil-20 data pre-processing

```python
coil_data_path = '/content/drive/MyDrive/Colab Notebooks/AutoML/datasets/coil-20-proc'

import os
import pandas as pd

coil_data = []
for root, dirs, files in os.walk(coil_data_path):
    for file in files:
        if file.endswith('.png'):
            label = int(file.split('__')[0].split('obj')[1])
            coil_data.append({'image': os.path.join(root, file), 'label': label})
coil_data = pd.DataFrame(coil_data)
print(coil_data.head())
print('shape:', coil_data.shape)
print('number of classes:', coil_data['label'].nunique())
```

```python
# Divide the dataset into training and testing sets
from sklearn.model_selection import train_test_split

coil_train_data, coil_test_data = train_test_split(coil_data, test_size=0.2, random_state=42)
coil_train_data.head()
```

## Default model: swin_base_patch4_window7_224

### with warm-starting

```python
from comet_ml import Experiment

experiment = Experiment(
  api_key="lWgl4dpDBFwIgiFoRl48gx5RI",
  project_name="automl-coil20",
  workspace="kittyysm"
)
```

```python
from autogluon.multimodal import MultiModalPredictor
from codecarbon import EmissionsTracker

with EmissionsTracker() as tracker:
    # Compute intensive training code goes here

    # model_path = '/content/drive/MyDrive/Colab Notebooks/AutoML/models-autogluon/coil-swin_base_patch4_window7_224'
    # predictor = MultiModalPredictor(label="label", path=model_path)
    predictor = MultiModalPredictor(label="label") # do not save the model

    predictor.fit(
        train_data = coil_train_data,
        hyperparameters = {"model.timm_image.checkpoint_name": "swin_base_patch4_window7_224",
                           "optimization.warmup_steps": 0.1 },
    )

# output accuracy on test set
scores = predictor.evaluate(coil_test_data, metrics=["accuracy"])
print('Top-1 test acc: %.3f' % scores["accuracy"])

# print training summary
summary = predictor.fit_summary()
print(summary)
```

```python
experiment.end()
```
