# Few Shot in plants30 dataset

## Load Dataset (only 10 images per class)

```python
import os
import pandas as pd

plant_data_path = '/content/drive/MyDrive/Colab Notebooks/AutoML/datasets/plants-30'

labels_map = {
    'aloevera': 0, 'banana': 1, 'bilimbi': 2, 'cantaloupe': 3, 'cassava': 4, 'coconut': 5, 'corn': 6,
    'cucumber': 7, 'curcuma': 8, 'eggplant': 9, 'galangal': 10, 'ginger': 11, 'guava': 12, 'kale': 13,
    'longbeans': 14, 'mango': 15, 'melon': 16, 'orange': 17, 'paddy': 18, 'papaya': 19, 'peperchili': 20,
    'pineapple': 21, 'pomelo': 22, 'shallot': 23, 'soybeans': 24, 'spinach': 25, 'sweetpotatoes': 26,
    'tobacco': 27, 'waterapple': 28, 'watermelon': 29
}

# Initialize the plant data list
plant_data = []

# Initialize a dictionary to keep track of counts per category
category_count = {label: 0 for label in labels_map.values()}

for root, dirs, files in os.walk(plant_data_path):
    for file in files:
        if file.endswith('.jpg'):
            category = root.split('/')[-1]
            label = labels_map[category]
            # Check if the current category has less than 10 images loaded
            if category_count[label] < 10:
                plant_data.append({'image': os.path.join(root, file), 'label': label})
                # Increment the count for this category
                category_count[label] += 1

# Convert the list to a DataFrame
plant_data_df = pd.DataFrame(plant_data)

print(plant_data_df.head())
print('shape:', plant_data_df.shape)
print('number of classes:', plant_data_df['label'].nunique())
```

```python
# Divide the dataset into training and testing sets
from sklearn.model_selection import train_test_split

plant_train_data, plant_test_data = train_test_split(plant_data_df, test_size=0.2, random_state=42)
plant_train_data.head()
```

## Train a Few Shot Classifier

```python
from comet_ml import Experiment

experiment = Experiment(
  api_key="lWgl4dpDBFwIgiFoRl48gx5RI",
  project_name="few-shot-plants30",
  workspace="kittyysm"
)
```

```python
from autogluon.multimodal import MultiModalPredictor

from codecarbon import EmissionsTracker
tracker = EmissionsTracker()
tracker.start()
try:
     # Compute intensive code goes here
     predictor_fs_image = MultiModalPredictor(
     problem_type="few_shot_classification",
     label="label",  # column name of the label
     eval_metric="acc",
     )
     predictor_fs_image.fit(plant_train_data)
     scores = predictor_fs_image.evaluate(plant_test_data, metrics=["acc", "f1_macro"])
     print(scores)

finally:
     tracker.stop()
```

```python
experiment.end()
```
